Exercício 1: Desenvolva um algoritmo que solicite ao usuário dois números inteiros e dois números float. Imprima a soma destes quatro números na tela.

```python
# Vamos somar alguns numeros
try:
  Valor_1 = int(input('Por favor digite o numero inteiro para a soma: '))
  Valor_2 = int(input('Por favor digite o numero inteiro para a soma: '))

except Exception as e:
  print('Ocorreu um erro: Insira apenas numeros inteiros ', e)

try:
  Valor_3 = float(input('Por favor digite o numero float para a soma: '))
  Valor_4 = float(input('Por favor digite o numero float para a soma: '))

except Exception as e:
  print('Ocorreu um erro: Insira apenas numeros float ', e)

Valor_total = Valor_1 + Valor_2 + Valor_3 + Valor_4

print('\t\nA soma total dos valores é = {}' .format(Valor_total))
```

Exercício 2: Desenvolva um algoritmo que solicite ao usuário uma quantidade de dias, horas, minutos e segundos. Calcule o total de segundos resultante e imprima na tela para o usuário.


```python
# Vamos contar os segundos de uma periodo de tempo

Dias = int(input('Digite a quantidade de dias: '))
Horas = int(input('Digite a quantidade de horas: '))
Minutos = int(input('Digite a quantidade de Minutos: '))
Segundos = int(input('Digite a quantidade de Segundos: '))

Dias_em_Segundos = Dias * 24 * 60 * 60
Horas_em_Segundos = Horas * 60 * 60
Minutos_em_Segundos = Minutos * 60


print('\nDias em Segundos = {}' .format(Dias_em_Segundos))
print('\nHoras em Segundos = {}' .format(Horas_em_Segundos))
print('\nMinutos em Segundos = {}' .format(Minutos_em_Segundos))
print('\nSegundos = {}' .format(Segundos))

Total_de_Segundos = Dias_em_Segundos + Horas_em_Segundos + Minutos_em_Segundos + Segundos

print('\nO Total de segudos é = {}' .format(Total_de_Segundos))
```

Exercício 3: Desenvolva um algoritmo que solicite ao usuário o preço de um produto e um percentual de desconto a ser aplicado a ele. Calcule e exiba o valor do desconto e o preço final do produto.

```python
# Vamos simular uma plataforma de calculo de desconto

print('\n Simule aqui o seu desconto!')

Valor_do_produto = float(input('Digite o valor do produto: '))
Percentual_de_desconto = float(input('Digite o percentual de desconto: '))
Valor_do_desconto = Valor_do_produto * (Percentual_de_desconto / 100)
Valor_final_do_produto = Valor_do_produto - Percentual_de_desconto

print(f'\nO Valor do Final  é {Valor_final_do_produto:.6f}')
print(f'\nO Valor do Desconto é {Valor_do_desconto:.6f}')
```
